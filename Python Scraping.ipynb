# Importing all important Libraries

```python
import pandas as pd
import numpy as np
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
from tqdm import tqdm
from urllib.parse import urljoin
```

# Webdriver to use Chrome browser to open different pages

```python
driver = webdriver.Chrome("/usr/lib/chromium-browser/chromedriver")
```

# Getting the links of all the pages in the gives site

```python
try:
    driver.get("https://www.amazon.in/s?rh=n%3A6612025031&fs=true&ref=lp_6612025031_sar")
    content = driver.page_source
    soup = BeautifulSoup(content)
    all_links = []
    #Getting the relative reference of the pages from the page slider and doing url join to get the absolute link
    new_link = [(urljoin('https://www.amazon.in/',link.get("href"))) for link in soup("a",href=True, attrs={'class':'s-pagination-item s-pagination-button'})]
    #Copying the next page urls in a list
    all_links.append(new_link[0])
    driver.get(all_links[0])
    content = driver.page_source
    soup = BeautifulSoup(content)
    for i in range(2, 262):
            new_link = [(urljoin('https://www.amazon.in/',link.get("href"))) for link in soup("a",href=True, attrs={'class':'s-pagination-item s-pagination-button'})]
            lennum = len(new_link)-1
            all_links.append(new_link[lennum])
            driver.get(new_link[lennum])
            content = driver.page_source
            soup = BeautifulSoup(content)
except IndexError as e:
        print(e)
```

# Getting the required details of all the products

```python
products=[] #List to store name of the product
prices=[] #List to store price of the product
ratings=[] #List to store rating of the product

driver.get("https://www.amazon.in/s?rh=n%3A6612025031&fs=true&ref=lp_6612025031_sar")
content = driver.page_source
soup = BeautifulSoup(content)
#Get all the sponsored products details
for a in soup.findAll('div', attrs={'class':'sg-col-4-of-24 sg-col-4-of-12 s-result-item s-asin sg-col-4-of-16 AdHolder sg-col s-widget-spacing-small sg-col-4-of-20'}):
    name=a.find('span', attrs={'class':'a-size-base-plus a-color-base a-text-normal'})
    products.append(name.text)
    rating=a.find('span', attrs={'class':'a-icon-alt'})
    ratings.append(rating.text)
    price=a.find('span', attrs={'class':'a-price-whole'})
    prices.append(price.text)
    
#Get all the other products details
for a in soup.findAll('div', attrs={'class':'sg-col-4-of-24 sg-col-4-of-12 s-result-item s-asin sg-col-4-of-16 sg-col s-widget-spacing-small sg-col-4-of-20'}):
    name=a.find('span', attrs={'class':'a-size-base-plus a-color-base a-text-normal'})
    products.append(name.text)
    rating=a.find('span', attrs={'class':'a-icon-alt'})
    ratings.append(rating.text)
    price=a.find('span', attrs={'class':'a-price-whole'})
    prices.append(price.text)
    
for i in range(0, len(all_links)):
    driver.get(all_links[i])
    content = driver.page_source
    soup = BeautifulSoup(content)
    for a in soup.findAll('div', attrs={'class':'sg-col-4-of-24 sg-col-4-of-12 s-result-item s-asin sg-col-4-of-16 AdHolder sg-col s-widget-spacing-small sg-col-4-of-20'}):
        name=a.find('span', attrs={'class':'a-size-base-plus a-color-base a-text-normal'})
        products.append(name.text)
        rating=a.find('span', attrs={'class':'a-icon-alt'})
        if rating == None:
            ratings.append(None)
        else:
            ratings.append(rating.text)
        price=a.find('span', attrs={'class':'a-price-whole'})
        if price == None:
            prices.append(None)
        else:
            prices.append(price.text)
    
        #Get all the other products details
    for a in soup.findAll('div', attrs={'class':'sg-col-4-of-24 sg-col-4-of-12 s-result-item s-asin sg-col-4-of-16 sg-col s-widget-spacing-small sg-col-4-of-20'}):
        name=a.find('span', attrs={'class':'a-size-base-plus a-color-base a-text-normal'})
        products.append(name.text)
        rating=a.find('span', attrs={'class':'a-icon-alt'})
        if rating == None:
            ratings.append(None)
        else:
            ratings.append(rating.text)
        price=a.find('span', attrs={'class':'a-price-whole'})
        if price == None:
            prices.append(None)
        else:
            prices.append(price.text)
    print(len(products),len(prices),len(ratings))
```

# Writing the details of all the products into CSV file

```python
df = pd.DataFrame({'Product Name':products,'Price':prices,'Rating':ratings}) 
df.to_csv('products.csv', index=False, encoding='utf-8')
```
